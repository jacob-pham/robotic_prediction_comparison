```python
import torch, matplotlib.pyplot as plt
data = torch.load('datasets_processed/eth/train.pt')
#lets look at 1 trajectory (20 frames, 2 coordinates)

traj = data[1000].numpy()
print(traj.shape)  # should be (20, 2)

#plot the trajectory
plt.plot(traj[:, 0], traj[:, 1], marker='o')
for i in range(traj.shape[0]):
    plt.text(traj[i, 0], traj[i, 1], str(i), fontsize=8)
plt.title('Example Trajectory')
```
